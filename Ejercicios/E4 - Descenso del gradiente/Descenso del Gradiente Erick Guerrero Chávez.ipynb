```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
c = pd.read_csv("Advertising.csv")
x = c.iloc[:,1]
y = c.iloc[:,4]
plt.scatter(x,y,c="blue",alpha=0.5)
```

```python
m = 0
c = 0

L = 0.00001 # Taza de aprendizaje.
epochs = 10000  # NÃºmero de iteraciones.

n = float(len(x))

for i in range(epochs): 
    Y_pred = m*x + c  # Valor predecido para Y
    D_m = (-2/n) * sum(x * (y - Y_pred))  # Derivada con respecto a M
    D_c = (-2/n) * sum(y - Y_pred)  # Derivada con respecto a C
    m = m - L * D_m
    c = c - L * D_c
    
print (m, c)
```

```python
Y_pred = m*x + c

plt.scatter(x, y) 
plt.plot([min(x), max(x)], [min(Y_pred), max(Y_pred)], color='red') 
plt.show()
```
