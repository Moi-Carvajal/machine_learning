## Ejercicio 5

Desarrollar un modelo de regresión lineal mediante el método de mínimos cuadrados ordinarios con el Dataset "Advertising.csv" (adjunto en la sección de abajo) que te permita calcular/predecir el Aumento en ventas de un producto en particular en base a lo invertido en la publicidad.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv("advertising.csv")
dc=pd.DataFrame(data,columns=["TV","Sales"])
x= np.array(dc["TV"])
y= np.array(dc["Sales"])
plt.scatter(x,y)
```

```python
m =0
c =0
L = 0.00001
ep = 1000

n = float(len(x))

for i in range(ep):
    Y = m * x + c
    Dm = (-2/n) * sum(x * (y - Y))
    Dc = (-2/n) * sum(y - Y)
    m = m - L * Dm
    c = c - L * Dc
    
print(m,c)
 
Y = m * x + c

plt.scatter(x,y)
plt.plot([min(x),max(x)],[min(Y),max(Y)],c="green")
plt.show
```
