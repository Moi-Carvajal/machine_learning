```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
c = pd.read_csv("Advertising.csv")
x = c.iloc[:,1]
y = c.iloc[:,4]
plt.scatter(x,y,c="blue",alpha=0.5)
```

```python
m = 0
c = 0

L = 0.00003                               #Taza de aprendizaje.
epochs = 150000                           #Iteraciones.

n = float(len(x))

for i in range(epochs): 
    pred_en_y = m*x + c                      #Predicción Y
    D_m = (-2/n) * sum(x * (y - pred_en_y))  #Derivada de M
    D_c = (-2/n) * sum(y - pred_en_y)        #Derivada de C
    m = m - L * D_m
    c = c - L * D_c
    
print (m, c)
```

```python
pred_en_y = m*x + c

plt.scatter(x, y) 
plt.plot([min(x), max(x)], [min(pred_en_y), max(pred_en_y)], color='black') 
plt.show()
```
