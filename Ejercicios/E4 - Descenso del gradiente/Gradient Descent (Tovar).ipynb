```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
adv=pd.read_csv("Advertising.csv")
x=np.array(adv['TV'])
y=np.array(adv['Sales'])
plt.scatter(x,y,c="orange",alpha=0.5)
plt.show()
```

```python

m=0
c=0
L=0.00001
ep=1000
n=len(x)
for i in range(n):
    ypred=m*x+c
    D_m=(-2/n) * sum(x*(y-ypred))
    D_c=(-2/n) * sum(y-ypred)
    m=m-L*D_m
    c=c-L*D_c
print(m,c)    
    
```
