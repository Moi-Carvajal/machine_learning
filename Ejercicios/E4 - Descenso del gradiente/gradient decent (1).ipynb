```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("Advertising.csv")
pf =pd.DataFrame(data, columns = ["TV","Sales"])
X = np.array(pf["TV"])
Y = np.array(pf["Sales"])

plt.scatter(X,Y,c="blue",alpha=0.5)
```

```python
m = 0
c = 0

Lrate =0.00001 
it = 1000

n = float(len(X)) 

 
for i in range(it): 
    Y_pred = m*X + c 
    D_m = (-2/n) * sum(X * (Y - Y_pred))  
    D_c = (-2/n) * sum(Y - Y_pred)  
    m = m - Lrate * D_m  
    c = c - Lrate * D_c  
    
print (m, c)
```

```python
Y_pred = m*X + c

plt.scatter(X, Y) 
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)],  color='green')  # regression line
plt.show()
```
