## Cargamos las librerias y el archivo .csv

```python
import numpy as np
from numpy.linalg import inv
import pandas as pd 
import matplotlib.pyplot as plt

Ad_df=pd.read_csv('Advertising.csv')
Ad_df
```

## Creamos una tabla con los datos de las columnas TV y Sales

```python
Df=pd.DataFrame(Ad_df,columns=['TV','Sales'])
Df
```

## Regresión lineal y graficación.

```python
x = np.array(Df['TV'])  ##Convertimos TV y Sales en arrays y los graficamos.
y = np.array(Df['Sales'])
plt.scatter(x,y)

x = np.array([np.ones(200),x]).T
b = inv(x.T @ x) @ x.T @ y  ##Formula para minimizar los cuadrados.
plt.plot([0,300], [b[0]+b[1]*0, b[0]+b[1]*300 ], c = 'black')
plt.xlabel("TV"); plt.ylabel("Sales")  ##Nombramos los ejes de nuestra gráfica.
plt.show()
```
