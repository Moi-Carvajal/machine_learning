```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Advertising.csv")
data
```

## Crear un dataframe con las columnas "TV" y "Sales" e imprimirlas en pantalla.

```python
df=pd.DataFrame(data,columns=["TV","Sales"])
df
```

## Regresión lineal

```python
X = np.array (df["TV"])
Y = np.array (df["Sales"])

plt.scatter (X, Y)


X = np.array ([np.ones(200), X]).T
B = np.linalg.inv(X.T @ X) @ X.T @ Y

plt.plot([0,300], [B[0] + B[1] * 0, B[0] + B[1] * 300], c="red")
plt.show()
```
