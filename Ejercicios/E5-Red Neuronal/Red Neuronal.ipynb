```
import matplotlib.pyplot as plt
import scipy as sc
import numpy as np
from sklearn.datasets import make_circles 
```

```
n=500
p=2
x,y=make_circles(n_samples=n,factor=0.4,noise=0.06)
plt.scatter(x[y==0,0],x[y==0,1],c="orange")
plt.scatter(x[y==1,0],x[y==1,1],c="red",alpha=0.5)
plt.axis("equal")
```
